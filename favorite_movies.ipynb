# ðŸŽ¬ Movie Ratings Analysis

This project analyzes my top 10 movies and visualizes their ratings.

### âœ… Step 1: Create Movie and Rating Lists

```python
movies = ["Inception", "Interstellar", "Avatar", "Avengers", "Titanic",
          "Dark Knight", "Joker", "Up", "Toy Story", "Coco"]

ratings = [9, 9, 8, 8, 7, 9, 8, 7, 8, 8]
```

### âœ… Step 2: Display Movies with Their Ratings

```python
for movie, rating in zip(movies, ratings):
    print(movie, " - ", rating)
```

```python
%matplotlib inline
import matplotlib.pyplot as plt
plt.figure(figsize=(10,5))
plt.bar(movies, ratings)
plt.xticks(rotation=45)
plt.ylabel("Rating")
plt.title("My Movie Ratings")
plt.show()
```

```python
max_rating = max(ratings)
min_rating = min(ratings)

print("Highest Rated:", movies[ratings.index(max_rating)])
print("Lowest Rated:", movies[ratings.index(min_rating)])
```
